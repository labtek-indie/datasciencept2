```python
from sklearn.linear_model import LinearRegression 
from sklearn.cross_validation import train_test_split
```

```python
train1 = data.drop(['price','date','id','zipcode'], axis=1)
labels = data['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size=.5, random_state=2)
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_test,y_test)
```
